1.查看网络接口信息(ifconfig/ip)

```python
ifconfig eth0   #显示eth0接口信息
ifconfig        #显示所有激活网络接口信息
ip addr     #查看所有接口ip地址
ip link     #查看所有接口链路层信息
ip route    #查看路由表
```

2.测试网络连通性(ping)<br>
ping [选项] 目标地址<br>
选项：-c N(指定发送N个数据包后停止)、-i(设置数据包发送间隔秒数)

```python
ping -c 4 8.8.8.8       #向google DNS服务器发送4数据包，测试连通性
```

3.查看网络连接状态(ss/netstat)<br>
查看系统网络连接、端口监听、进程占用端口等信息

```python
netstat -tuln       #查看所有监听的TCP/UDP端口
#-t：显示TCP连接、-u：显示UDP连接、-l：只显示监听状态端口、-n以数字形式显示ip和端口
netstat -tulnp      #额外显示占用端口的进程PID
```

```python
ss -tuln    #同netstat -tuln 
ss -tulnp   #同netstat -tulnp
```

4.下载文件(wget/curl)<br>
从网络下载文件(支持HTTP、HTTPS、FTP协议)<br>
wget:专注下载<br>
curl: 功能更全

```python
wget https://example.com/file.zip  # 下载文件到当前目录
wget -c https://example.com/large.iso  # 断点续传（如果之前下载中断）
wget -O newname.zip https://example.com/file.zip  # 重命名下载的文件为 newname.zip
```

```python
curl -O https://example.com/file.zip    #保留原文件名下载
curl -o newname https://example.com/file.zip    #重命名文件下载
curl https://example.com  # 直接显示目标 URL 的内容（如网页源码）
```

5.远程登录(ssh)<br>
通过SSH协议远程登录到其他Linux服务器<br>
ssh 用户名@目标IP/域名 or ssh [选项] 用户名@目标IP/域名<br>
选项：-p 端口号(指定非默认端口)、-i 密钥文件(使用SSH密钥登录)

```python
ssh lisi@192.168.1.100
ssh -p 2222 root@example.com  # 登录 example.com 的 2222 端口（root 账号）
ssh -i ~/.ssh/mykey.pem ubuntu@ec2-xx-xx-xx-xx.compute-1.amazonaws.com  # 用密钥登录 AWS 服务器
```

6.传输文件(scp/sftp)<br>
本地和远程服务器之间传输文件(基于SSH协议)<br>
(1) scp: 简单文件传输<br>
本地->远程服务器:scp 本地文件 远程用户@远程IP:远程路径<br>
远程服务器->本地:scp 远程用户@远程IP:远程文件 本地路径<br>
常用选项：-r(递归传输文件)

```python
scp note.txt lisi@192.168.1.100:/data   #将本地note.txt传到服务器data目录中
scp -r ./note lisi@192.168.1.100:/data    #递归传输note目录
scp lisi@192.168.1.100:note.txt ./data    #将服务器note.txt传到本地data目录
```

(2) sftp:交互式文件传输<br>
sftp 远程用户@远程IP

```python
sftp lisi@192.168.1.100     #连接服务器，进入交互式页面
```

下面是一些常用交互命令：

```python
put local_file      #上传本地文件到远程当前目录
get remote_file     #下载远程文件到本地当前目录
put -r local_dir    #上传本地目录
get -r remote_dir   #下载远程目录
ls          #查看远程目录
lls         #查看本地目录
cd 远程路径     #切换远程目录
lcd 本地路径    #切换本地目录
exit        #退出
```

7.查看DNS配置(nslookup/dig)<br>
查询域名对应的IP地址<br>
nslookup: 简单；dig: 详细

```python
nslookup google.com     #查询google.com的IP地址
dig google.com          #显示google.com的详细DNS信息
```

8.配置临时网络(ifconfig/ip)<br>
临时修改网络接口的IP地址

```python
sudo ifconfig eth0 192.168.1.200 netmask 255.255.255.0      #临时设置eth0的IP地址
sudo ip addr add 192.168.1.200/24 dev eth0  # /24 等价于子网掩码 255.255.255.0
sudo ifconfig eth0 up   #激活eth0
sudo ifconfig eth0 down     #禁用eth0
```
