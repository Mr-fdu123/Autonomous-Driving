2.ls(list列出目录信息)

```python
ls  #列出当前目录的文件
```

```python
ls /data    #列出data目录文件
```

```python
ls -l /data     #列出data目录的详细信息
```

```python
ls -la /data    #列出data目录所有文件（含隐藏文件）
```

3.cd(change directory切换目录)

```python
cd /data    #切换到data目录
```

```python
cd ./data   #切换到当前目录下data目录
```

```python
cd ..   #回到上级目录
```

```python
cd -    #切换到上一次所在目录
```

4.-h or --help(显示命令用法)

```python
ls --help
```

5.mkdir(make directory创建目录)

```python
mkdir data  #创建名为data的目录
```

```python
mkdir -p a/b/c  #递归创建a,b,c三个目录
```

```python
mkdir ~/data    #在home目录下创建data目录
```

6.touch(创建文件)

```python
touch data.txt  #在当前目录下创建data.txt文件
```

```python
touch ~/data.txt    #在home目录下创建data.txt文件
```

7.cp(copy 复制文件或者目录)
cp [选项] 文件/目录 目标路径

```python
cp note.txt ./data/     #复制note.txt文件到data目录中
```

```python
cp -i a.txt b.txt   #如果b.txt文件已存在，则会提示是否覆盖
```

```python
cp -r model ./data/     #递归地将model目录复制到data目录下
```

```python
cp -v *.txt ./data/     #复制当前所有txt文件到data目录，并显示复制过程
```

8.mv(move 移动or重命名文件/目录)
mv [选项] 文件/目录 目标路径

```python
mv note.txt ./data/     #移动note.txt文件到data目录中
```

```python
mv oldname.txt newname.txt  # 把 oldname.txt 重命名为 newname.txt
```

```python
mv -i a.txt ./data/     #移动a.txt文件到data目录中，已存在a.txt文件已存在，则会提示是否覆盖
```

```python
mv data1 data2 ./data/      #同时移动data1和data2到data目录中
```

9.rm(remove 删除文件/目录)
rm [选项] 文件/目录

```python
rm note.txt     #删除note.txt文件
```

10.查看文件内容(多种方式)

```python
cat note.txt       # 显示 note.txt 的全部内容
cat a.txt b.txt    # 同时显示 a.txt 和 b.txt 的内容（拼接输出）
```

```python
head -5 README.md       #查看README前五行
```

```python
tail -3 README.md       #查看README后三行
```

```python
tail -f /nginx/access.log       #实时查看nginx访问日志
```
