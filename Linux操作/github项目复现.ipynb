### 展示一些github项目复现时常用的Linux命令

1.拉取github上的代码文件, 分两种情况：<br>
(1)服务器能翻墙，直接

```python
git clone https://github.com/仓库名
```

(2)服务器不能翻墙，需要本地下载zip文件，用git bash传输到服务器上，具体指令<br>
scp /本地文件路径 用户名@服务器ip:服务器目标目录路径

```python
(bash) scp /c/users/lenovo/downloads/file.zip root@192.168.1.100:/data/
```

通常还要解压缩到当前目录

```python
unzip file.zip
```

```python
unzip file -d /data     #解压缩到指定的data目录
```

2.编辑文件，环境变量

```python
nano train.sh       #编辑训练脚本
```

```python
nano ~/.bashrc      #编辑环境变量
source ~/.bashrc        #使环境变量生效
```

3.创建，激活环境

```python
conda create --name myenv python=3.9    #创建并命名环境
conda activate myenv    #激活环境
conda deactivate        #退出环境
```
