### 一、文件权限基础

Linux系统中，文件/目录的权限分为三类对象：<br>
1.所有者User, u:创建文件/目录的用户<br>
2.所属组Group, g:user所在的用户组(可包含多个用户)<br>
3.其他用户Other, o:既不是user, 也不是group的用户<br>
每个对象拥有三个权限:<br>
1.Read, r:读，权限值4，允许查看文件内容或者列出目录中的文件<br>
2.Write, w:写，权限值2，允许修改文件内容或者在目录中增删文件(删改文件权限取决于目录权限)<br>
3.Execute, x:执行，权限值1，允许运行可执行文件或者进入目录

### 二、查看文件权限

```python
ls -l note.txt
# 输出：-rw-r--r-- 1 user group 1234 10月  5 14:30 note.txt
```

看看输出：第一个字符表示文件类型(-普通文件，d目录，l链接文件)<br>
接下来9个字符表示文件权限分组(每三个一组，分别表示u, g, o)<br>
1.前三个：user权限(可读，可写，不可执行)<br>
2.中三个：group权限(可读，不可写，不可执行)<br>
3.后三个：other权限(可读，不可写，不可执行)

### 三、修改文件权限

1.数字法(三个权限之和表示权限，本质上转换为二进制即可，
例如7二进制数为111，表示rwx；5=101，表示r-x)<br>
chmod [权限数字] 文件/目录

```python
chmod 755 train.sh      #user权限rwx，group和other权限为r-x
```

2.符号法(符号操作对象和权限)<br>
对象：u, g, o, a(all 对象)<br>
操作：+添加权限，-移除权限，=设置权限

```python
chmod u+x train.sh      #user添加执行权限
chmod a-w train.sh      #所有对象移除可写权限
```

### 四、修改所有者

用于改变user(需要管理员权限sudo)<br>
sudo chown new_user 文件/目录

```python
sudo chown lisi note.txt    
sudo chown lisi:dev /data       #data目录user改为lisi，group改为dev
```

### 五、修改所属组

单独修改文件/目录所属组(需要管理员权限sudo)<br>
sudo chgrp new_group 文件/目录

```python
sudo chgrp dev train.sh     #将train.sh所属组改为dev
```

### 六、用户管理(需要管理员权限)

Linux 通过用户账号区分不同操作者，每个用户有独立的家目录（/home/用户名）和权限

1.创建用户(useradd)<br>
sudo useradd [选项] user_name<br>
选项有-m(自动创建用户home目录)、-s(指定默认shell，通常为bash)

```python
sudo useradd -m -s /bin/bash lisi   #创建用户lisi，默认shell为bash
```

2.设置用户密码(passwd)

```python
sudo passwd lisi
passwd         #当前用户修改密码
```

3.删除用户<br>
sudo userdel [选项] user_name
选项有-r(删除用户及home目录、邮件目录)

```python
sudo userdel -r lisi
```

### 七、用户组管理(需要管理员权限)

1.创建用户组(groupadd)，删除用户组(groupdel)

```python
sudo groupadd dev   #创建dev组
sudo groupdel dev   #删除dev组
```

2.用户加入组(usermod)

```python
sudo usermod -aG dev lisi   #将lisi加入dev组，-a表示追加，-G表示指定附加组
```

4.查看用户所属组

```python
group   #显示当前用户所属组
groups lisi     #显示lisi所属组
cat /etc/group      #查看系统所有用户组
```
